```python
# 1. IMPORT LIBRARY
import pandas as pd
import os

# 2. DEFINISI FOLDER DAN LOKASI DATASET (Sesuai dengan direktori Colab Anda)
folder_path = "Dataset_Amazon"
file_name = "amazon.csv"

# 3. PENGECEKAN FOLDER LOKAL (Pengganti proses download)
if os.path.exists(folder_path):
    print("Folder dataset berhasil ditemukan di Colab")
else:
    raise Exception("Gagal menemukan folder dataset")

# 4. PENGECEKAN FILE CSV (Pengganti proses ekstrak ZIP)
file_path = os.path.join(folder_path, file_name)

if os.path.exists(file_path):
    print("File CSV tersedia di dalam folder")
else:
    raise Exception("File amazon.csv tidak ditemukan")

# 5. BACA FILE CSV MENGGUNAKAN PANDAS
df = pd.read_csv(file_path, sep=',')
print("Dataset berhasil dibaca ke DataFrame")
```

```python
# 13. FEATURE ENGINEERING & MAPPING DATA
# (Pengganti mapping bulan karena Amazon.csv tidak memiliki kolom waktu)

import numpy as np

print("\n=== MAPPING PRIORITAS KINERJA PRODUK ===\n")

# Convert 'rating' column to numeric, coercing errors to NaN
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# 1. Membuat kategori teks berdasarkan KPI 4 dan KPI 5 Anda
kondisi = [
    (df['rating'] >= 4.0),                       # KPI 5: Superior
    (df['rating'] >= 3.0) & (df['rating'] < 4.0), # Normal
    (df['rating'] < 3.0)                         # KPI 4: Buruk
]
pilihan_label = ['Superior', 'Normal', 'Buruk']

# Memasukkan kondisi di atas ke dalam kolom baru
df['performance_label'] = np.select(kondisi, pilihan_label, default='Tidak Diketahui')

# 2. PROSES MAPPING (Sama persis seperti konsep template Anda)
# Kita mapping status teks tadi menjadi angka prioritas untuk manajemen
# 1 = Mendesak (harus dicabut), 2 = Pantau, 3 = Aman (fokus diiklankan)
priority_mapping = {
    'Buruk': 1,
    'Normal': 2,
    'Superior': 3
}

# Terapkan mapping ke kolom baru
df['action_priority'] = df['performance_label'].map(priority_mapping)

# 3. Cek hasilnya (Menampilkan 10 data teratas)
display(df[['product_id', 'rating', 'performance_label', 'action_priority']].head(10))
```

**BUSINESS METRIC & DATA DICTIONARY**

```python
# 14. BUSINESS METRIC: DEFINISI TARGET & KPI DASAR
# (Mengganti target 'y/yes' menjadi target produk 'Superior')

print("\n=== BUSINESS METRIC: TINGKAT KESUKSESAN KATALOG PRODUK ===\n")

# 1. Membuat target binary (1/0).
# Jika rating >= 4.0 dianggap 'YES' (1), selain itu 'NO' (0)
df['is_superior'] = (df['rating'] >= 4.0).astype(int)

# 2. Kalkulasi Metrik Dasar (Sama seperti template conversion rate)
total_records = len(df)
total_superior = df['is_superior'].sum()

# Menghitung persentase kesuksesan (Superiority Rate / KPI 5)
superior_rate = total_superior / total_records

# 3. Menampilkan Hasil
print("Total Seluruh Produk di Etalase  :", total_records)
print("Total Produk Superior (YES/Sukses):", total_superior)
print("Tingkat Kesuksesan (KPI 5)       :", round(superior_rate * 100, 2), "%")
```

```python
# 16. FEATURE ENGINEERING UNTUK DASHBOARD AMAZON
# (Membagi angka menjadi kelompok/kategori agar mudah divisualisasikan)

print("\n=== PENGELOMPOKAN DATA (FEATURE ENGINEERING) ===\n")

# --- Preprocessing: Convert relevant columns to numeric types ---
# Convert 'discount_percentage' from object to float
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '', regex=False).astype(float)

# Convert 'discounted_price' from object to float (remove '₹' and commas)
df['discounted_price'] = df['discounted_price'].str.replace('₹', '', regex=False).str.replace(',', '', regex=False).astype(float)

# Convert 'actual_price' from object to float (remove '₹' and commas)
df['actual_price'] = df['actual_price'].str.replace('₹', '', regex=False).str.replace(',', '', regex=False).astype(float)

# Convert 'rating_count' from object to float (remove commas)
df['rating_count'] = df['rating_count'].str.replace(',', '', regex=False).astype(float)

# 1. Discount Bucket (Mengelompokkan persentase diskon - Terkait KPI 3)
df['discount_bucket'] = pd.cut(
    df['discount_percentage'],
    bins=[-1, 10, 30, 60, 100],  # Batas angka persentase
    labels=['Promo Kecil (<=10%)', 'Promo Sedang (11-30%)', 'Promo Besar (31-60%)', 'Cuci Gudang (>60%)'],
    include_lowest=True
)

# 2. Price Category (Membagi kelas harga menjadi 4 tingkat menggunakan kuartil - Terkait KPI 8)
# qcut akan membagi data sama rata berdasarkan jumlah produknya
df['price_category'] = pd.qcut(
    df['discounted_price'],
    q=4,
    labels=['Q1: Murah / Entry-level', 'Q2: Menengah', 'Q3: Menengah-Atas', 'Q4: Premium / Mahal']
)

# 3. Engagement Level (Mengelompokkan tingkat keramaian pembeli - Terkait KPI 2 & 9)
df['engagement_level'] = pd.cut(
    df['rating_count'],
    bins=[-1, 500, 5000, 20000, 10000000],
    labels=['Sepi (<500 ulasan)', 'Normal (500-5K)', 'Ramai (5K-20K)', 'Viral / Laris (>20K)'],
    include_lowest=True
)

# 4. Tampilkan hasil kolom asli bersanding dengan kolom baru hasil pengelompokan
kolom_tampil = [
    'product_id',
    'discount_percentage', 'discount_bucket',
    'discounted_price', 'price_category',
    'rating_count', 'engagement_level'
]

display(df[kolom_tampil].head(10))
```

```python
# 17. EDA BERBASIS KPI: Tingkat Kesuksesan (Superior Rate) Per Segmen
# (Mengganti 'job' dengan 'main_category')

import matplotlib.pyplot as plt

print("\n=== TINGKAT KESUKSESAN PRODUK (SUPERIOR RATE) BERDASARKAN KATEGORI ===\n")

# 1. Pastikan kolom kategori utama (main_category) sudah terbentuk
# (Memotong teks kategori Amazon yang panjang, ambil kata pertamanya saja)
if 'main_category' not in df.columns:
    df['main_category'] = df['category'].str.split('|').str[0]

# 2. Membuat Fungsi Pembuatan Tabel KPI
def segment_kpi(col, top_n=12):
    # Menggunakan target 'is_superior' (dari Langkah 14) sebagai pengganti 'y_bin'
    tmp = df.groupby(col).agg(
        total=('is_superior', 'count'),       # Total produk di kategori tersebut
        superior_yes=('is_superior', 'sum'),  # Jumlah produk yang ratingnya >= 4.0
        superior_rate=('is_superior', 'mean') # Persentase kesuksesan produknya
    ).sort_values('superior_rate', ascending=False)

    # Hapus kategori yang produknya terlalu sedikit (misal < 5 produk) agar adil
    tmp = tmp[tmp['total'] >= 5]
    return tmp.head(top_n)

# 3. Menghitung metrik berdasarkan Kategori Utama - Top 5 Kategori Terbaik
category_kpi = segment_kpi('main_category', 5)
print("Tabel Top 5 Kategori dengan Superior Rate Tertinggi:")
display(category_kpi)

# 4. Membuat Visualisasi Grafik Batang (Bar Chart)
plt.figure(figsize=(10, 5))
# Mengalikan superior_rate dengan 100 agar tampilannya berwujud persen (0-100%)
bars = plt.bar(category_kpi.index.astype(str), category_kpi['superior_rate'].values * 100, color='royalblue')

plt.title('Top 5 Kategori dengan Kinerja Produk Terbaik (KPI 5)', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Persentase Superior (%)')
plt.ylim(0, 110) # Memberi ruang kosong di atas grafik agar teks persentase tidak terpotong

# Menambahkan label angka persentase tepat di atas setiap batang grafik
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 2, f'{yval:.1f}%', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# 18. EDA BERBASIS KPI: Analisis Multi-Segmen Otomatis
# (Mengganti kolom bank dengan kolom Feature Engineering Amazon)

print("\n=== ANALISIS KESUKSESAN PRODUK BERDASARKAN BERBAGAI SEGMEN ===\n")

# Daftar kolom yang ingin dianalisis beserta jumlah baris yang ingin ditampilkan
# Format: ('nama_kolom', jumlah_data)
fitur_analisis = [
    ('discount_bucket', 4),   # Menganalisis pengaruh besaran diskon
    ('price_category', 4),    # Menganalisis pengaruh kelas harga
    ('engagement_level', 4)   # Menganalisis pengaruh keramaian ulasan
]

# Perulangan (loop) untuk membuat tabel dan grafik dari masing-masing fitur di atas
for col, n in fitur_analisis:

    # 1. Panggil fungsi segment_kpi (dari langkah 17)
    tmp = segment_kpi(col, n)

    print(f"\n---> Tabel Kinerja Berdasarkan Segmen: {col.upper()}")
    display(tmp)

    # 2. Buat Grafiknya
    plt.figure(figsize=(8, 4.5))

    # Mengambil nilai superior_rate dan dikalikan 100 agar jadi persen
    bars = plt.bar(tmp.index.astype(str), tmp['superior_rate'].values * 100, color='blue')

    plt.title(f'Tingkat Kesuksesan (Superior Rate) Berdasarkan {col}', fontsize=12, fontweight='bold')
    plt.xticks(rotation=45, ha='right')
    plt.ylabel('Superior Rate (%)')
    plt.ylim(0, 115) # Memberikan ruang kosong di atas grafik agar teks tidak terpotong

    # Tambahkan angka persentase tepat di atas setiap batang grafik
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 2, f'{yval:.1f}%', ha='center', va='bottom', fontweight='bold')

    plt.tight_layout()
    plt.show()
```

```python
# 19. TREN KESUKSESAN PRODUK (Mengganti tren bulanan dengan tren diskon)
# (Dashboard Trend Line Chart)

import matplotlib.pyplot as plt
import pandas as pd

print("\n=== TREN TINGKAT KESUKSESAN BERDASARKAN BESARAN DISKON ===\n")

# 1. Menghitung agregasi (Sama seperti template, menggunakan discount_bucket)
# observed=False ditambahkan agar Pandas tidak memunculkan peringatan (warning)
trend_kpi = df.groupby('discount_bucket', observed=False).agg(
    total=('is_superior', 'count'),
    superior_yes=('is_superior', 'sum'),
    superior_rate=('is_superior', 'mean')
).reset_index()

# 2. Mengurutkan data (Karena discount_bucket sudah dibuat terurut di Langkah 16,
# kita tinggal memanggil sort_values agar urutannya pasti dari Kecil -> Besar)
trend_kpi = trend_kpi.sort_values('discount_bucket')

print("Tabel Tren Kinerja Berdasarkan Besaran Diskon:")
display(trend_kpi)

# 3. Membuat Grafik Garis (Line Chart)
plt.figure(figsize=(10, 5))

# Menggambar garis dan titik (marker='o'), nilai dikali 100 agar jadi persen
plt.plot(trend_kpi['discount_bucket'].astype(str), trend_kpi['superior_rate'] * 100,
         marker='o', markersize=8, color='green', linewidth=2.5)

plt.title('Tren Kesuksesan (Superior Rate) Seiring Besarnya Diskon', fontsize=14, fontweight='bold')
plt.xlabel('Kategori Diskon', fontsize=12)
plt.ylabel('Superior Rate (%)', fontsize=12)
plt.ylim(0, 115) # Ruang kosong agar angka di atas titik tidak terpotong
plt.grid(True, linestyle='--', alpha=0.5) # Menambahkan garis bantu (grid) agar lebih estetik

# 4. Menambahkan label angka persentase tepat di atas setiap titik
for i in range(len(trend_kpi)):
    yval = trend_kpi['superior_rate'].iloc[i] * 100
    xval = trend_kpi['discount_bucket'].astype(str).iloc[i]
    plt.text(xval, yval + 3, f'{yval:.1f}%', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# 20. KINERJA MARKETING: Pengaruh Detail Deskripsi Terhadap Kesuksesan
# (Mengganti 'Duration Campaign' dengan 'Panjang Deskripsi Produk')

import matplotlib.pyplot as plt
import pandas as pd

print("\n=== ANALISIS KESUKSESAN BERDASARKAN DETAIL DESKRIPSI PRODUK ===\n")

# 1. Feature Engineering: Menghitung jumlah karakter di kolom deskripsi produk
df['desc_length'] = df['about_product'].astype(str).str.len()

# 2. Membuat 'Bucket' (Kelompok) untuk panjang deskripsi
df['desc_bucket'] = pd.cut(
    df['desc_length'],
    bins=[0, 500, 1000, 2000, 50000], # Batas karakter
    labels=['Singkat (<500 kar)', 'Sedang (500-1K kar)', 'Detail (1K-2K kar)', 'Sangat Detail (>2K kar)'],
    include_lowest=True
)

# 3. Menghitung metrik agregasi kesuksesan (Superior Rate)
# parameter observed=False untuk menghindari warning dari Pandas versi terbaru
desc_kpi = df.groupby('desc_bucket', observed=False).agg(
    total=('is_superior', 'count'),
    superior_yes=('is_superior', 'sum'),
    superior_rate=('is_superior', 'mean')
).reset_index()

print("Tabel Kinerja Berdasarkan Tingkat Detail Deskripsi:")
display(desc_kpi)

# 4. Membuat Visualisasi Grafik Batang (Bar Chart)
plt.figure(figsize=(8, 4.5))

# Plot bar chart, nilai dikalikan 100 agar menjadi persentase
bars = plt.bar(desc_kpi['desc_bucket'].astype(str), desc_kpi['superior_rate'] * 100, color='blue')

plt.title('Superior Rate Berdasarkan Tingkat Detail Deskripsi', fontsize=13, fontweight='bold')
plt.xlabel('Kelompok Panjang Teks Deskripsi', fontsize=11)
plt.ylabel('Superior Rate (%)', fontsize=11)
plt.ylim(0, 115) # Memberi ruang di atas grafik agar angka tidak terpotong

# Menambahkan label angka persentase tepat di atas setiap batang grafik
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 2, f'{yval:.1f}%', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# 21. KINERJA KATALOG: Pengaruh Panjang Nama Produk (SEO) Terhadap Kesuksesan
# (Mengganti 'Campaign Frequency' dengan 'Jumlah Kata Nama Produk')

import matplotlib.pyplot as plt

print("\n=== TREN KESUKSESAN BERDASARKAN PANJANG NAMA PRODUK (SEO) ===\n")

# 1. Feature Engineering: Menghitung jumlah kata pada nama produk
df['name_word_count'] = df['product_name'].astype(str).str.split().str.len()

# 2. Melakukan 'Clipping' (Membatasi nilai maksimal seperti di template)
# Jika ada nama produk yang panjangnya 30 atau 50 kata, kita gabungkan di kelompok "20" (Maksimal 20 kata)
df['name_word_clip'] = df['name_word_count'].clip(upper=20)

# 3. Menghitung metrik agregasi kesuksesan (Superior Rate)
name_kpi = df.groupby('name_word_clip').agg(
    total=('is_superior', 'count'),
    superior_yes=('is_superior', 'sum'),
    superior_rate=('is_superior', 'mean')
).reset_index().sort_values('name_word_clip')

print("Tabel Kinerja Berdasarkan Panjang Kata di Judul Produk (Maks 20):")
display(name_kpi.head(10)) # Menampilkan 10 data teratas

# 4. Membuat Visualisasi Grafik Garis (Line Chart)
plt.figure(figsize=(9, 5))

# Plot grafik garis dengan titik (marker='o'), nilai dikali 100 agar jadi persen
plt.plot(name_kpi['name_word_clip'], name_kpi['superior_rate'] * 100,
         marker='o', color='purple', linewidth=2.5, markersize=7)

plt.title('Tren Superior Rate Berdasarkan Panjang Nama Produk', fontsize=14, fontweight='bold')
plt.xlabel('Jumlah Kata pada Judul Produk (Dibatasi Maks 20)', fontsize=12)
plt.ylabel('Superior Rate (%)', fontsize=12)
plt.ylim(0, 115) # Memberi ruang kosong di atas grafik
plt.grid(True, linestyle='--', alpha=0.5)

# Opsi tambahan: Menambahkan label persentase pada titik tertentu agar tidak terlalu penuh
for i in range(0, len(name_kpi), 2): # Hanya menampilkan angka selang-seling agar rapi
    yval = name_kpi['superior_rate'].iloc[i] * 100
    xval = name_kpi['name_word_clip'].iloc[i]
    plt.text(xval, yval + 3, f'{yval:.0f}%', ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# 22. KINERJA SPESIFIK: Pengaruh Sub-Kategori Terhadap Kualitas Produk
# (Mengganti 'Previous Outcome' dengan 'Sub-Kategori')

import matplotlib.pyplot as plt

print("\n=== ANALISIS TINGKAT KESUKSESAN BERDASARKAN SUB-KATEGORI ===\n")

# 1. Feature Engineering: Mengambil data Sub-Kategori
# Kategori Amazon dipisah dengan '|'. Contoh: "Electronics|HomeTheater|TV"
# str[0] adalah kategori utama, str[1] adalah sub-kategori. Kita ambil elemen ke-2 (index 1).
df['sub_category'] = df['category'].str.split('|').str[1]

# 2. Menghitung metrik agregasi kesuksesan (Superior Rate)
subcat_kpi = df.groupby('sub_category').agg(
    total=('is_superior', 'count'),
    superior_yes=('is_superior', 'sum'),
    superior_rate=('is_superior', 'mean')
).sort_values('superior_rate', ascending=False)

# 3. Filter Data Bias
# Kita hanya ambil sub-kategori yang memiliki minimal 10 produk agar hitungannya adil,
# lalu kita potong hanya untuk menampilkan Top 7 Sub-Kategori terbaik.
subcat_kpi = subcat_kpi[subcat_kpi['total'] >= 10].head(7)

print("Tabel Top 7 Sub-Kategori dengan Superior Rate Tertinggi:")
display(subcat_kpi)

# 4. Membuat Visualisasi Grafik Batang (Bar Chart)
plt.figure(figsize=(10, 5))

# Plot bar chart, nilai dikalikan 100 agar menjadi persentase
bars = plt.bar(subcat_kpi.index.astype(str), subcat_kpi['superior_rate'].values * 100, color='darkorange')

plt.title('Top 7 Sub-Kategori Produk Paling Memuaskan (Superior Rate)', fontsize=14, fontweight='bold')
plt.ylabel('Superior Rate (%)', fontsize=12)
plt.ylim(0, 115) # Memberi ruang agar teks persentase tidak terpotong
plt.xticks(rotation=20, ha='right', fontsize=10) # Memiringkan teks agar nama sub-kategori tidak bertabrakan

# Menambahkan label angka persentase tepat di atas setiap batang grafik
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 2, f'{yval:.1f}%', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```
